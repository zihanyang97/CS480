```python
import csv
from sklearn.neural_network import MLPRegressor
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold

# function to import train data
def importData(file):
    data = []
    csvFile = open(file, "r")
    csv_reader = csv.reader(csvFile, delimiter=',')
    
    for row in csv_reader:
        tmp = []
        for col in row:
            tmp.append(float(col))
        data.append(tmp)
        
    return data

def nn(k, X, x, Y, y):
    clf = MLPRegressor(solver='sgd', hidden_layer_sizes=(k), activation='logistic')
    clf.fit(X, Y)
    pred = clf.predict(x)
    mse = 0
    for i in range(len(x)):
        mse += float(pred[i] - y[i])**2
    return mse / len(x)

# import train input & target
X = []
Y = []
for i in range(1, 11):
    X += importData("trainInput" + str(i) + ".csv")
    Y += importData("trainTarget" + str(i) + ".csv")
X = np.array(X)
Y = np.array(Y)

# import test input & target
x = importData("testInput.csv")
y = importData("testTarget.csv")

# 10-fold validation initialize
kf = KFold(n_splits=10)

unitsRange = np.arange(1, 11)
minMSE = float('inf')
score_list = []
bestunits = 0

for i in unitsRange:
    total = 0
    for train_index, test_index in kf.split(X):
        #print("train:", train_index, "validation:", test_index)
        X_train, X_test = X[train_index], X[test_index]
        Y_train, Y_test = Y[train_index], Y[test_index]
        total += nn(i, X_train, X_test, Y_train, Y_test) 
    # compute avgMSE
    avgMSE = total / 10
    if avgMSE < minMSE:
        minMSE = avgMSE
        bestunits = i
    print("when degree = " + str(i))
    print("MSE for Gaussian Process with polynomial is " + str(avgMSE))
    score_list.append(avgMSE)
    
    
print("best number of hidden units is " + str(bestunits))

result = nn(bestunits, X, x, Y, y)

print("MSE using best degree " + str(bestunits) + " for test set is " + str(result))

plt.plot(unitsRange, score_list)
plt.xlabel("number of hidden units in neural network")
plt.ylabel("MSE")
plt.show()
```
