```python
import numpy as np
from numpy import *
import csv
import math
import operator
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
from sklearn.preprocessing import PolynomialFeatures


# function to import train data
def importData(file):
    data = []
    csvFile = open(file, "r")
    csv_reader = csv.reader(csvFile, delimiter=',')
    
    for row in csv_reader:
        tmp = []
        for col in row:
            tmp.append(float(col))
        data.append(tmp)
        
    return data

def gaussianProcess(sig, xTrain, xValidation, yTrain, yValidation):
    MSE = 0
    # get I 
    I = np.zeros((len(xTrain),len(xTrain)))
    for i in range(len(I)):
        I[i][i] = 1
    # get gram matrix
    K = np.zeros((len(xTrain),len(xTrain)))
    X = np.matrix(xTrain)
    X2 = np.matrix(xValidation)
    for i in range(len(xTrain)):
        for j in range(len(xTrain)):
            K[i,j] = exp(-0.5 * dot(X[i] - X[j],(X[i] - X[j]).getT()) / sig**2)    
    k = np.zeros((len(xValidation), len(xTrain)))
    for i in range(len(xValidation)):
        for j in range(len(xTrain)):
            k[i,j] = exp(-0.5 * dot(X2[i] - X[j], (X2[i] - X[j]).getT()) / sig**2)
    # compute
    K = K + I
    K = np.matrix(K)
    KI = K.getI()
    pred = dot(k, dot(KI, yTrain))
    for i in range(len(xValidation)):
        MSE += float(pred[i] - yValidation[i])**2
    return MSE / len(xValidation)

# import train input & target
X = []
Y = []
for i in range(1, 11):
    X += importData("trainInput" + str(i) + ".csv")
    Y += importData("trainTarget" + str(i) + ".csv")
X = np.array(X)
Y = np.array(Y)

# import test input & target
x = importData("testInput.csv")
y = importData("testTarget.csv")

# 10-fold validation initialize
kf = KFold(n_splits=10)

sigmaRange = np.arange(1, 7)
minMSE = float('inf')
score_list = []
bestsig = 0

for i in sigmaRange:
    total = 0
    for train_index, test_index in kf.split(X):
        #print("train:", train_index, "validation:", test_index)
        X_train, X_test = X[train_index], X[test_index]
        Y_train, Y_test = Y[train_index], Y[test_index]
        total += gaussianProcess(i, X_train, X_test, Y_train, Y_test)   
    # compute avgMSE
    avgMSE = total / 10
    if avgMSE < minMSE:
        minMSE = avgMSE
        bestsig = i
    print("when sigma = " + str(i))
    print("MSE for Gaussian Process with gaussian is " + str(avgMSE))
    score_list.append(avgMSE)
    

print("best sigma is " + str(bestsig))

result = gaussianProcess(bestsig, X, x, Y, y)

print("MSE using best sigma " + str(bestsig) + " for test set is " + str(result))

plt.plot(sigmaRange, score_list)
plt.xlabel("value of sigma for gaussian process with gaussian kernel")
plt.ylabel("MSE")
plt.show()
```
