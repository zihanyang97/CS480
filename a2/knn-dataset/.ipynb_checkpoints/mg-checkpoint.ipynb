```python
import numpy as np
from numpy import *
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import csv
import math
import random
import operator
import matplotlib.pyplot as plt
import collections

# function to import train data
def importData(file):
    data = []
    csvFile = open(file, "r")
    csv_reader = csv.reader(csvFile, delimiter=',')
    
    for row in csv_reader:
        tmp = []
        for col in row:
            tmp.append(int(col))
        data.append(tmp)
        
    return data

def ls(a):
    return 1 / (1 + math.exp(-a))

# import train data & labels
X = []
Y = []
for i in range(1, 11):
    X += importData("trainData" + str(i) + ".csv")
    Y += importData("trainLabels" + str(i) + ".csv")
X = np.array(X)
Y = np.array(Y)

#import test data & labels
x = importData("testData.csv")
y = importData("testLabels.csv")

N1 = 0
N2 = 0
for i in Y:
    if i[0] == 5:
        N1 += 1
    elif i[0] == 6:
        N2 += 1

pi = N1 / len(Y)

prob1 = pi
prob2 = 1 - pi

MU1 = np.zeros((64, 1))
MU2 = np.zeros((64, 1))

for i in range(len(X)):
    tmp = np.matrix(X[i])
    tmp = tmp.getT()    
    if Y[i][0] == 5:
        MU1 += tmp
    elif Y[i][0] == 6:
        MU2 += tmp

MU1 = np.matrix(MU1 / N1)
MU2 = np.matrix(MU2 / N2)


S1 = np.zeros((64, 64))
S2 = np.zeros((64, 64))

for i in range(len(X)):
    tmp = np.matrix(X[i])
    tmp = tmp.getT()    
    if Y[i][0] == 5:
        tmp = tmp - MU1
        tmp = np.dot(tmp, tmp.getT())
        S1 += tmp
    elif Y[i][0] == 6:
        tmp = tmp - MU2
        tmp = np.dot(tmp, tmp.getT())    
        S2 += tmp

S1 = S1 / N1
S2 = S2 / N2

sig = np.matrix(S1 * N1/len(X) + S2 * N2/len(X))

print("pi is :" + str(pi))
print("MU1 is: " + str(MU1))
print("MU2 is: " + str(MU2))
print("diagonal of big sigma is: " + str(sig.diagonal()))


W = dot(sig.I, MU1-MU2)
T = dot(sig.getI(), MU1-MU2)
W0 = -0.5 * dot(dot(MU1.getT(), sig.I), MU1) + 0.5 * dot(dot(MU2.getT(), sig.I), MU2) + np.log(prob1/prob2)

corr = 0
for i in range(len(x)):
    pr = ls(dot(W.getT(), x[i]) + W0)
    if pr >= 0.5:
        if y[i][0] == 5:
            corr += 1
    else:
        if y[i][0] == 6:
            corr += 1
print("the accuracy is: " + str(corr / len(x)))
```
