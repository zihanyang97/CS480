```python
import csv
import numpy as np
from numpy import *
import math
import random
import operator
from sklearn.model_selection import KFold
import matplotlib.pyplot as plt

# each train file has 20 instances
# with 2 attributes

# function to import train data
def importData(file):
    data = []
    csvFile = open(file, "r")
    csv_reader = csv.reader(csvFile, delimiter=',')
    
    for row in csv_reader:
        tmp = []
        for col in row:
            tmp.append(col)
        data.append(tmp)
        
    return data

# function for linear regression
def linearRegression(k, xTrain, xValidation, yTrain, yValidation):
    # set mse to infinite large value
    MSE = 0
    # get lambdaI
    I = np.matrix([[1,0,0],[0,1,0],[0,0,1]])
    lambdaI = k * I
    # get A inverse
    M = np.matrix(xTrain)
    M = M.getT()
    A = dot(M,M.getT()) + lambdaI
    AI = A.getI()
    # get b
    b = zeros((3,1))
    for i in range(0,len(xTrain)):
        N = np.matrix(xTrain[i])
        N = N.getT()
        b += float(yTrain[i][0]) * N
    W = dot(AI,b)
    #print W
    W0 = float(W[0])
    W1 = float(W[1])
    W2 = float(W[2])
    #print W # check W correctness
    for i in range(0, len(xValidation)):
        tmp = np.matrix(xValidation[i])
        tmp = tmp.getT()        
        predict = float(dot(W.getT(), tmp))
        MSE += (predict - float(yValidation[i][0]))**2
    return MSE / len(xValidation)

# import train data
X = []
X += importData("trainInput1.csv")
X += importData("trainInput2.csv")
X += importData("trainInput3.csv")
X += importData("trainInput4.csv")
X += importData("trainInput5.csv")
X += importData("trainInput6.csv")
X += importData("trainInput7.csv")
X += importData("trainInput8.csv")
X += importData("trainInput9.csv")
X += importData("trainInput10.csv")

# add 1 into every X
for i in X:
    i.insert(0, 1)
    i[1] = float(i[1])
    i[2] = float(i[2])
X = np.array(X)

# import train labels
Y = []
Y += importData("trainTarget1.csv")
Y += importData("trainTarget2.csv")
Y += importData("trainTarget3.csv")
Y += importData("trainTarget4.csv")
Y += importData("trainTarget5.csv")
Y += importData("trainTarget6.csv")
Y += importData("trainTarget7.csv")
Y += importData("trainTarget8.csv")
Y += importData("trainTarget9.csv")
Y += importData("trainTarget10.csv")
Y = np.array(Y)

#print len(trainInput)
#print len(trainTarget)


# import test data
x = importData("testInput.csv")

for i in x:
    i.insert(0, 1)
    i[1] = float(i[1])
    i[2] = float(i[2])

# import test labels
y = importData("testTarget.csv")

# set lambda
lambda_range = np.arange(0.1, 4.1, 0.1)
minMSE = float('inf')
score_list = []
K = 0

# 10-fold validation initialize
kf = KFold(n_splits=10)

for k in lambda_range:
    MSE = 0
    for train_index, test_index in kf.split(X):
        #print("train:", train_index, "validation:", test_index)
        X_train, X_test = X[train_index], X[test_index]
        Y_train, Y_test = Y[train_index], Y[test_index]
        MSE += linearRegression(k, X_train, X_test, Y_train, Y_test)
        
    avgMSE = MSE / 10
    if avgMSE < minMSE:
        K = k
        minMSE = avgMSE
    score_list.append(avgMSE)

print("---------------result----------------")
print("best lambda is " + str(K))

result = linearRegression(K, X, x, Y, y)

print("MSE using best lambda " + str(K) + " is " + str(result))

plt.plot(lambda_range, score_list)
plt.xlabel("value of lambda for linear regression")
plt.ylabel("MSE")
plt.show()
```
